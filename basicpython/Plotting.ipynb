```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5, 5, 0.01)
y1 = -5*x*x + x + 10
y2 = 5*x*x + x

fig, ax = plt.subplots()
ax.plot(x, y1, x, y2, color='black')
#fill between
ax.fill_between(x, y1, y2, where=(y1 < y2), facecolor='red', alpha=0.1)

ax.vlines(x=0, ymin=-120, ymax=120, colors='red', linestyles='dashed', label='zero')

ax.axhspan(ymin=-100, ymax=-50, xmin=0, xmax=.9, alpha=0.1, color='grey')
```

```python
from matplotlib.ticker import MultipleLocator, FuncFormatter
np.random.seed(1)
x = np.random.randn(1000).cumsum()
y = np.random.randn(1000).cumsum()

fig, ax = plt.subplots(num=1)
ax.plot(x)
ax.plot(y, color='green')
ax.xaxis.set_major_locator(MultipleLocator(100))
ax.xaxis.set_minor_locator(MultipleLocator(50))

def format_hundred(x, pos):
    if x % 200 == 0:
        return str(x)
    else:
        return None

ax.xaxis.set_major_formatter(FuncFormatter(format_hundred))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position(('outward', 10))
ax.spines['left'].set_position(('outward', 10))
plt.grid(True, which='major')
plt.grid(True, which='minor', color='blue', linestyle='--')
ax.set_title(r"Random numbers $\sum^{1000}_{i}X_{i}$, $X_{i} \textasciitilde N(0, 1)$")
plt.annotate("Cumulative sum \n of normal variables",(600, 40), (0, 40), arrowprops= dict(arrowstyle = '->') , color='red')
fig.set_size_inches(11, 8)
plt.fill_between(range(1000), x, y, (x > y), color='yellow', alpha = 0.5)
```
