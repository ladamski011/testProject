# Graphwiz

````python
from graphviz import Digraph

graph = Digraph(engine='dot')
graph.attr('graph', rankdir='LR')
graph.attr('node', fontname='Handlee', shape='diamond')
graph.attr('edge', fontname='Handlee', arrowhead='vee')

graph.node('1', "One")
graph.attr('node', fontname='Handlee', shape='box')
graph.node('2', "Two")
graph.node('4', "Three")
graph.edge('1', '4')
graph.edge('4', '3')
graph.edges(['12', '21'])
graph.edge('2', '3', label="Noted")
graph.node('3', "Final")
graph.render('myTest1.gv')
graph
````
